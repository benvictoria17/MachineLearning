```python
import numpy as np 
import pandas as pd
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/benvictoria17/MachineLearning/master/dataset/Video%20Game%20Sales%20with%20Ratings/Video_Games_Sales_as_at_22_Dec_2016.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.drop('Other_Sales', axis = 1) # apaga a coluna
dataset = dataset.drop('Global_Sales', axis = 1) # apaga a coluna
dataset = dataset.drop('Developer', axis = 1) # apaga a coluna
dataset.info()
```

```python
dataset = dataset.dropna(axis = 0) # apagar por linha
dataset.info()
```

```python
valores_menores_1 = dataset.loc[dataset['NA_Sales'] < 1]
valores_menores_1.head()
```

```python
dataset = dataset.loc[dataset['NA_Sales'] > 1]
dataset = dataset.loc[dataset['EU_Sales'] > 1]
dataset.info()
```

```python
nome_jogos = dataset['Name']
dataset = dataset.drop('Name', axis = 1) # Assim, vamos diminuir a variedade de dados
dataset.info()
```

```python
from keras.layers import Dense, Dropout, Activation, Input
from keras.models import Model
previsores = dataset.iloc[:, [0,1,2,3,7,8,9,10,11]].values
previsores
```

```python
vendas_na = dataset.iloc[:, 4].values
vendas_eu = dataset.iloc[:, 5].values
vendas_jp = dataset.iloc[:, 6].values
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
# Transformar colunas categóricas em numéricas

labelEncoder = LabelEncoder()

previsores[:, 0] = labelEncoder.fit_transform(previsores[:, 0]);
previsores[:, 2] = labelEncoder.fit_transform(previsores[:, 2]);
previsores[:, 3] = labelEncoder.fit_transform(previsores[:, 3]);
previsores[:, 8] = labelEncoder.fit_transform(previsores[:, 8]);
previsores
```

```python

column_transform = ColumnTransformer([("encoder", 
                         OneHotEncoder(), 
                        [0,2,3,8])],    
                       remainder = 'passthrough')


previsores = column_transform.fit_transform(previsores).toarray()
previsores
```

```python
previsores.shape
```

```python
quantidade_saida = 3
quantidade_entrada = previsores.shape[1]
units = (quantidade_entrada + quantidade_saida) / 2
units
```

```python

camada_entrada = Input(shape=(61,))

camada_oculta_1 = Dense(units = units, activation = 'sigmoid')(camada_entrada) # ligação entre as camadas
camada_oculta_2 = Dense(units = units, activation = 'sigmoid')(camada_oculta_1)

# Previsão de três saídas para as três regiões previstas
camada_saida_1 = Dense(units = 1, activation = 'linear')(camada_oculta_2)
camada_saida_2 = Dense(units = 1, activation = 'linear')(camada_oculta_2)
camada_saida_3 = Dense(units = 1, activation = 'linear')(camada_oculta_2)

regressor = Model(inputs = camada_entrada, 
                  outputs = [camada_saida_1, camada_saida_2, camada_saida_3])

regressor.compile(optimizer = 'adam', loss = 'mse')

regressor.fit(previsores, [vendas_na, vendas_eu, vendas_jp],
             epochs = 50, batch_size = 100)
```

```python
previsao_na, previsao_eu, previsao_jp = regressor.predict(previsores)
previsao_na[0:10]
```

```python
vendas_na[0:10]
```

```python
features_final = np.column_stack((nome_jogos, vendas_na, previsao_na, vendas_eu, previsao_eu, vendas_jp, previsao_jp))
features_final
```

```python
dataset_previsao = pd.DataFrame(data = features_final, 
                                columns = ['Names', 
                                           'NA_Sales',
                                           'NA_Sales_Predict',
                                           'EU_Sales',
                                           'EU_Sales_Predict',
                                           'JP_Sales',
                                           'JP_Sales_Predict'])
dataset_previsao.head(20)
```
