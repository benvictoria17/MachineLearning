```python
x2 = x.drop('diagnosis', axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x1 = pd.DataFrame(scaler.fit_transform(x2))
x1.columns = x2.columns
```

```python
x1.head()
```

```python
y1 = y
plt.figure(figsize=(10,10))
data = pd.concat([y, x1.iloc[:,0:10]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='values')
sns.swarmplot(x='features', y='values', hue='diagnosis',  data=data)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,10))
data = pd.concat([y, x1.iloc[:,10:20]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='values')
sns.swarmplot(x='features', y='values', hue='diagnosis',  data=data)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,10))
data = pd.concat([y, x1.iloc[:,20:31]], axis=1)
data = pd.melt(data, id_vars='diagnosis',var_name='features',value_name='values')
sns.swarmplot(x='features', y='values', hue='diagnosis',  data=data)
plt.xticks(rotation=90)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x2, y, test_size=0.2)
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
rfe = RFE(estimator=clf, n_features_to_select=5, step=1)
rfe = rfe.fit(X_train,y_train)
print('Features which are selected: ', X_train.columns[rfe.support_])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
acs = accuracy_score(y_test,rfe.predict(X_test))
print('Accuracy is: ', acs)
cm = confusion_matrix(y_test,rfe.predict(X_test))
sns.heatmap(cm, annot=True)
```

```python
from sklearn.feature_selection import RFECV
rfecv = RFECV(estimator=clf, step=1, cv=5, scoring='accuracy')
rfecv = rfecv.fit(X_train,y_train)
print('Optimal number of features: ', rfecv.n_features_)
print('Features which are selected: ', X_train.columns[rfecv.support_])
```

```python
acs = accuracy_score(y_test,rfecv.predict(X_test))
print('Accuracy is: ', acs)
cm = confusion_matrix(y_test,rfecv.predict(X_test))
sns.heatmap(cm, annot=True)
```

```python
X_train1 = X_train[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'radius_se', 'area_se', 'symmetry_se',
       'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst',
       'smoothness_worst', 'compactness_worst', 'concavity_worst',
       'concave points_worst', 'symmetry_worst']]
X_test1 = X_test[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'radius_se', 'area_se', 'symmetry_se',
       'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst',
       'smoothness_worst', 'compactness_worst', 'concavity_worst',
       'concave points_worst', 'symmetry_worst']]
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn = knn.fit(X_train1,y_train)
acs = accuracy_score(y_test,knn.predict(X_test1))
print('Accuracy is: ', acs)
cm = confusion_matrix(y_test,knn.predict(X_test1))
sns.heatmap(cm, annot=True)
```
